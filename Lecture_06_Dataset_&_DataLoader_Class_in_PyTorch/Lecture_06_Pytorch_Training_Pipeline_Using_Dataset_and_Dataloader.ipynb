```python
# Import necessary libraries
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
# Load the dataset into a pandas DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')

df.head()
```

```python
# Display the shape of the DataFrame
df.shape
```

```python
# Drop unnecessary columns from the DataFrame
df.drop(columns=['id','Unnamed: 32'], inplace= True)
```

```python
# Display the first few rows of the updated DataFrame
df.head()
```

# Train test split

```python
# Perform train-test split on the dataset
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

# Scaling

```python
# Scale the features using StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Display the scaled training features
X_train
```

```python
# Display the training labels
y_train
```

**Label Encoding**

```python
# Perform label encoding on the target variable
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
# Display the encoded training labels
y_train
```

**Numpy arrays to Pytorch tensors**

```python
# Convert numpy arrays to PyTorch tensors
X_train = torch.from_numpy(X_train.astype(np.float32))
X_test = torch.from_numpy(X_test.astype(np.float32))
y_train = torch.from_numpy(y_train.astype(np.float32))
y_test = torch.from_numpy(y_test.astype(np.float32))
```

```python
# Display the shape of the training features tensor
X_train.shape
```

```python
# Display the shape of the training labels tensor
y_train.shape
```

```python
# Define a custom PyTorch Dataset class
from torch.utils.data import Dataset, DataLoader

class CustomDataset(Dataset):
  def __init__(self, features, labels):
    self.features = features
    self.labels = labels

  def __len__(self):

    return len(self.features)

  def __getitem__(self, idx):

    return self.features[idx], self.labels[idx]
```

```python
# Create instances of the custom dataset for training and testing
train_dataset = CustomDataset(X_train, y_train)
test_dataset = CustomDataset(X_test, y_test)
```

```python
# Access an example item from the training dataset
train_dataset[10]
```

```python
# Create DataLoader instances for training and testing
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=True)
```

**Defining the Model**

```python
# Define the neural network model
import torch.nn as nn

class MySimpleNN(nn.Module):
  def __init__(self, num_features):
    super().__init__()
    self.linear = nn.Linear(num_features, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):

    out = self.linear(features)
    out = self.sigmoid(out)

    return out
```

# Important Parameters

```python
# Define important training parameters
learning_rate = 0.1
epochs = 25
```

```python
# Initialize the model, optimizer, and loss function
model = MySimpleNN(X_train.shape[1])

optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

loss_funcion = nn.BCELoss()
```

#**Training Pipeline**

```python
# Training loop for the neural network
for epoch in range(epochs):

  for batch_features, batch_labels in train_loader:

    y_pred = model(batch_features)

    loss = loss_funcion(y_pred, batch_labels.view(-1, 1))

    optimizer.zero_grad()

    loss.backward()

    optimizer.step()

  print(f'Epoch: {epoch}, Loss: {loss.item()}')
```

# Evaluation

```python
# Evaluate the trained model's accuracy
model.eval()

# set the model to evaluation mode
accuracy_list = []

with torch.no_grad():
  for batch_features, batch_labels in test_loader:

    # forward pass
    y_pred = model(batch_features)
    y_pred = (y_pred > 0.5).float()

    # calculate accuracy for the current batch
    batch_accuracy = (y_pred.view(-1) == batch_labels).float().mean().item()
    accuracy_list.append(batch_accuracy)

# calculate overall accuracy
overall_accuracy = sum(accuracy_list) / len(accuracy_list)

print(f'Overall Accuracy: {overall_accuracy:.4f}')
```
