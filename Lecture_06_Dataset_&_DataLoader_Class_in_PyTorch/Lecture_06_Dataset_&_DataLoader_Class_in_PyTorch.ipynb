```python
# Import necessary libraries
from sklearn.datasets import make_classification
import torch
```

```python
# Generate synthetic classification data
X, y = make_classification(
    n_samples=10,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_classes=2,
    random_state=42
)
```

```python
# Display the features (X)
X
```

```python
# Display the shape of the features (X)
X.shape
```

```python
# Display the labels (y)
y
```

```python
# Display the shape of the labels (y)
y.shape
```

```python
# Convert NumPy arrays to PyTorch tensors
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.long)
```

```python
# Display the PyTorch features (X)
X
```

```python
# Display the PyTorch labels (y)
y
```

```python
# Import PyTorch data utilities
from torch.utils.data import Dataset, DataLoader
```

```python
# Define a custom PyTorch Dataset class
class CustomDataset(Dataset):
  def __init__(self, features, labels):
    self.features = features
    self.labels = labels

  def __len__(self):
    return self.features.shape[0]

  def __getitem__(self, index):
    return self.features[index], self.labels[index]
```

```python
# Create an instance of the custom dataset
dataset = CustomDataset(X, y)
```

```python
# Get the length of the dataset
len(dataset)
```

```python
# Access an item from the dataset
dataset[2]
```

```python
# Create a DataLoader for batching data
dataloader = DataLoader(dataset, batch_size=2, shuffle=False)
```

```python
# Iterate and print batches from the DataLoader
for batch_features, batch_labels in dataloader:
  print(batch_features)
  print(batch_labels)
  print("-"*50)
```
