```python
# Import necessary libraries
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
# Load the dataset
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
```

```python
# Display the first few rows of the DataFrame
df.head()
```

```python
# Display the shape of the DataFrame
df.shape
```

```python
# Drop unnecessary columns
df.drop(columns=['id', 'Unnamed: 32'], inplace=True)
```

```python
# Display the DataFrame after dropping columns
df.head()
```

# train test split

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0], test_size=0.2)
```

#Scaling

```python
# Scale the features using StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
#
```

```python
# Display the scaled training features
X_train
```

```python
# Display the training labels
y_train
```

# Label Encoding

```python
# Encode target labels
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
# Display the encoded training labels
y_train
```

# Numpy arrays to Pytorch tensors

```python
# Convert numpy arrays to PyTorch tensors
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
# Display the shape of the training features tensor
X_train_tensor.shape
```

```python
# Display the shape of the training labels tensor
y_train_tensor.shape
```

#**Defining the model**

```python
# Define a simple neural network class
class MySimpleNN():
  def __init__(self, X):
    self.weights = torch.rand(X.shape[1], 1, dtype=torch.float64,
    requires_grad=True)
    self.bias = torch.zeros(1, dtype=torch.float64, requires_grad=True)

  def forward(self, X):
    z = torch.matmul(X, self.weights) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred

  def loss_function(self, y_pred, y):
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1-epsilon)

    # calculate loss
    loss = -(y_train_tensor * torch.log(y_pred) + (1 - y_train_tensor) * torch.log(1 - y_pred)).mean()
    return loss
```

# Important Parameters

```python
# Define important training parameters
learning_rate = 0.1
epochs = 25
```

# Training Pipeline

```python
# Training pipeline for the neural network
model = MySimpleNN(X_train_tensor)

for epoch in range(epochs):

  y_pred = model.forward(X_train_tensor)

  loss = model.loss_function(y_pred, y_train_tensor)

  loss.backward()

  with torch.no_grad():
    model.weights -= learning_rate * model.weights.grad
    model.bias -= learning_rate * model.bias.grad

    model.weights.grad.zero_()
    model.bias.grad.zero_()

    print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
```

```python
# Display the trained model's bias
model.bias
```

# Evaluation

```python
# Evaluate the model's accuracy on the test set

with torch.no_grad():
  y_pred = model.forward(X_test_tensor)
  y_pred = (y_pred > 0.9).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
