```python
import torch
print(torch.__version__)
```

# Creating a Tensor

```python
# using empty
a = torch.empty(2,3)
print(a)
```

```python
# check type
type(a)
```

```python
# using zeros
b = torch.zeros(2,3)
print(b)
```

```python
# using ones
c = torch.ones(2,3)
print(c)
```

```python
# using rand
torch.rand(2,3)
```

```python
# use of seed
torch.rand(2,3)
```

```python
# manual_seed
torch.manual_seed(100)
torch.rand(2,3)
```

```python
torch.manual_seed(100)
torch.rand(2,3)
```

```python
# using tensor
torch.tensor([[1,2,3],[4,5,6]])
```

```python
# other ways

# arange

print('using arange ->', torch.arange(0,10,2))

# linspace

print('using linspace ->', torch.linspace(0,10,10))

# using eye
print('using eye ->', torch.eye(5))

# using full
print('using full ->', torch.full((2,3), 5))

```
